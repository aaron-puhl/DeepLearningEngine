```python
from DeepLearningEngine.backpropagation import Value
from DeepLearningEngine.nn import MLP

import numpy as np

data = [
    [Value(1),Value(1)],
    [Value(1),Value(-1)],
    [Value(-1),Value(1)],
    [Value(-1),Value(-1)],
]

targets = [Value(0),Value(1),Value(1),Value(0)]

```

```python
nn = MLP(2, [4,4,1])
```

```python
loss = None
for i in range(100):
    if loss is not None:
        loss.zero_grad()

    preds = [nn.forward(x) for x in data]
    loss = np.sum([(x - y)**2 for (x,y) in zip(targets, preds)]) 
    loss.backwards()

    print("loss:",loss)

    for param in nn.params():
        param._value = param._value - 0.01 * param.gradient
```
